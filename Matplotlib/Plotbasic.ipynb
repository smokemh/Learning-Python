# Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Making Plots Actually Look Nice

```python
plt.style.use(['science', 'notebook', 'grid'])
```

## Quick Plotting

- Simple plotting commands to look at data quickly

```python
x = np.linspace(0,15,30)
y = np.sin(x) + 0.1*np.random.randn(len(x))
```

- Normal Plot

```python
plt.plot(x,y)
```

- Dashed Lines

```python
plt.plot(x,y,"--")
```

- Points of the functions

```python
plt.plot (x,y,"o")
```

- Points with the Dashed Lines

```python
plt.plot(x,y,"o--")
```

- Change Color

```python
plt.plot(x,y,"o--",color= "red")
```

- Line Width 

```python
plt.plot (x,y,"o--", color = "red", lw=1)
```
